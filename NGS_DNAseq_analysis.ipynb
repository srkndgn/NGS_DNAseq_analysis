## Install bcftools

```python
conda install bcftools
```

```python
!conda config --show channels
```

```python
# Calculate the read coverage of positions in the genome
# -O b means output file in bcf format
# -o directs output to a specific folder
# -f tags path to reference genome
```

```python
!bcftools mpileup -O b -f reference/ecoli_rel606.fasta \
results/bam/SRR2589044.aligned.sorted.bam \
-o results/bcf/SRR2589044_raw.bcf
```

```python
# Calling the variants
# ploidy - 1 for bacteria
# -m allows for multiallelic and rare-variant calling
# -v tells the program to output variant sites only
```

```python
!bcftools call --ploidy 1 -m -v results/bcf/SRR2589044_raw.bcf \
-o results/vcf/SRR2589044_variants.vcf
```

```python
# Checking the raw variant file
```

```python
!head results/vcf/SRR2589044_variants.vcf
```

```python
# Filtering variants
```

```python
!vcfutils.pl varFilter results/vcf/SRR2589044_variants.vcf > results/vcf/SRR2589044_final_variants.vcf
```

```python
!head results/vcf/SRR2589044_final_variants.vcf
```

```python
!tail results/vcf/SRR2589044_final_variants.vcf
```
